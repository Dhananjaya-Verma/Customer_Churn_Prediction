```python
import os, joblib, pandas as pd
models_dir = os.path.abspath(os.path.join('..','models'))
pipeline = joblib.load(os.path.join(models_dir,'churn_pipeline.pkl'))

data_path = os.path.abspath(os.path.join(r"D:\CSE(DataScience)\Customer_Churn_Prediction\Datasets\Telco-Customer-Churn-dataset.csv"))
df = pd.read_csv(data_path)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0)
if 'customerID' in df.columns:
    df = df.drop('customerID', axis=1)
```

```python
X = df.drop('Churn', axis=1)
sample = X.iloc[:5].copy()
preds = pipeline.predict(sample)
probs = pipeline.predict_proba(sample)[:,1]

out = sample.copy()
out['pred_churn'] = preds
out['prob_churn'] = probs
out
```

```python
out.to_csv(os.path.abspath(os.path.join('..','outputs','sample_predictions.csv')), index=False)
print("Saved to outputs/sample_predictions.csv")
```
